```python
import pandas as pd
from jiwer import wer, cer, mer, wil, wip

# Load CSVs
df_small_e3 = pd.read_csv("evaluation_results_small_e3.csv")
df_small_e20 = pd.read_csv("evaluation_results_small_e20.csv")
df_large_e3 = pd.read_csv("evaluation_results_large_e3.csv")
df_large_e20 = pd.read_csv("evaluation_results_large_e20.csv")

# Function to compute ASR metrics
def evaluate_model(df):
    ref = df["Reference"].astype(str).tolist()
    hyp = df["Prediction"].astype(str).tolist()
    return {
        "WER": wer(ref, hyp),
        "CER": cer(ref, hyp),
        "MER": mer(ref, hyp),
        "WIL": wil(ref, hyp),
        "WIP": wip(ref, hyp)
    }

# Run evaluation
results = {
    "Whisper Small Epoch 3": evaluate_model(df_small_e3),
    "Whisper Small Epoch 20": evaluate_model(df_small_e20),
    "Whisper Large Epoch 3": evaluate_model(df_large_e3),
    "Whisper Large Epoch 20": evaluate_model(df_large_e20)
}

# Create DataFrame to display
df_results = pd.DataFrame(results).T
print(df_results.round(4))
```
