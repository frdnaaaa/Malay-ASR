```python
import pandas as pd
from jiwer import wer, cer, mer, wil, wip

# Load using comma separator instead of tab
df = pd.read_csv("transcription_results.csv")  # comma is default, no need to set sep

# Show column names to verify
print(df.columns.tolist())
```

```python
from jiwer import wer, cer, mer, wil, wip

# Drop rows with missing values
df_clean = df.dropna(subset=["Text Chunk", "small_transcription", "large_transcription"])

# Convert each column to list of strings
ref = df_clean["Text Chunk"].tolist()
hyp_small = df_clean["small_transcription"].tolist()
hyp_large = df_clean["large_transcription"].tolist()

# Compute metrics
metrics = {
    "WER": [wer(ref, hyp_small), wer(ref, hyp_large)],
    "CER": [cer(ref, hyp_small), cer(ref, hyp_large)],
    "MER": [mer(ref, hyp_small), mer(ref, hyp_large)],
    "WIL": [wil(ref, hyp_small), wil(ref, hyp_large)],
    "WIP": [wip(ref, hyp_small), wip(ref, hyp_large)],
}

# Create DataFrame for results
results_df = pd.DataFrame(metrics, index=["Small Model", "Large Model"])
print(results_df)
```
